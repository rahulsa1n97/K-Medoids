```python
c1=[]
c2=[]
c3=[]
c=[]
e=[]
k4=[6,2]
k3=[8,5]
a=[]
b=[]
for i in range(0,10):
    dist3=math.sqrt((k3[0]-data1.height[i])*(k3[0]-data1.height[i])) + math.sqrt((k3[1]-data1.weight[i])*(k3[1]-data1.weight[i]))
    dist4=math.sqrt((k4[0]-data1.height[i])*(k4[0]-data1.height[i])) + math.sqrt((k4[1]-data1.weight[i])*(k4[1]-data1.weight[i]))
    c1.append(dist3)
    c2.append(dist4)
print(c1)
print(c2)
d=random.randint(0,9)
#print(d)
for i in range(0,10):
    if(c1[i]<c2[i]):
        a.append(i)
    elif(c1[i]>c2[i]):
        b.append(i)
    i=d
print(i)
k4=[data1.height[i],data1.weight[i]]
#k4=d
print(k4)
print("Cost of k1 and k2 cluster:",sum(c1+c2))
print("Cluster k1 has the points:",a)
print("Cluster k2 has the points:",b)
for i in range(0,10):
    dist5=math.sqrt((k4[0]-data1.height[i])*(k4[0]-data1.height[i])) + math.sqrt((k4[1]-data1.weight[i])*(k4[1]-data1.weight[i]))
    c3.append(dist5)
print(c3)
for i in range(0,10):
    if(c1[i]<c3[i]):
        c.append(i)
    elif(c1[i]>c3[i]):
        e.append(i)
print("Cost of k1 and k2 cluster after choose a new cluster:",sum(c1+c3))
print("Cluster k1 has the points:",c)
print("Cluster k2 has the points:",e)
if(sum(c1+c2)<=sum(c1+c3)):
    print("No need to change in the Cluster")
else:
    print("Last Cluster classification was not good")
```
